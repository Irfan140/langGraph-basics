```python
from langgraph.graph import StateGraph, START, MessagesState
from langchain_groq import ChatGroq
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model = ChatGroq(model="moonshotai/kimi-k2-instruct-0905")
```

```python
# MessagesState is by default state, so no need to code for state
def call_model(state: MessagesState):
    response = model.invoke(state["messages"])
    return {"messages": [response]}
```

```python
builder = StateGraph(MessagesState)
builder.add_node("call_model", call_model)
builder.add_edge(START, "call_model")
```

```python
graph = builder.compile()
graph
```

```python
graph.invoke({"messages": [{"role": "user", "content": "Hi! My name is Irfan."}]})
```

```python
graph.invoke({"messages": [{"role": "user", "content": "What is my name?"}]})
```

```python
def call_model(state: MessagesState):
    response = model.invoke(state["messages"])
    return {"messages": [response]}
```

```python
builder = StateGraph(MessagesState)
builder.add_node("call_model", call_model)
builder.add_edge(START, "call_model")
```

```python
from langgraph.checkpoint.memory import InMemorySaver

checkpointer = InMemorySaver()
graph = builder.compile(checkpointer=checkpointer)
```

```python
config1 = {"configurable": {"thread_id": "thread-1"}}
config2 = {"configurable": {"thread_id": "thread-2"}}
```

```python
graph.invoke({"messages": [{"role": "user", "content": "Hi! My name is Irfan."}]}, config1)
```

```python
graph.invoke({"messages": [{"role": "user", "content": "What is my name?"}]}, config1)
```

```python
# Checking the messages present in the state
snap = graph.get_state(config1)
vals = snap.values
for m in vals.get("messages", []):
        print("-", type(m).__name__, ":", m.content)
```

```python
graph.invoke({"messages": [{"role": "user", "content": "Hi! I am learning Japanese."}]}, config2)
```

```python
graph.invoke({"messages": [{"role": "user", "content": "Hey what i am learning?"}]}, config2)
```

```python
# Checking the messages present in the state
snap = graph.get_state(config2)
vals = snap.values
for m in vals.get("messages", []):
        print("-", type(m).__name__, ":", m.content)
```
