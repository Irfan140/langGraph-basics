### Short Term Meory
- A conversation with a chatbot is a single session
- we maintain different short term meory for different conversations
- STM is thread scope
- STM is also called conversation buffer

```python
# add whole conversation (also called as thread) in context windows 
messages = ["My name is Irfan"]
```

```python
output = llm.invoke(messages).content
output
```

```python
messages.append(output)
messages
```

```python
messages.append("wht is my name?")
```

```python
output = llm.invoke(messages).content
output
```
